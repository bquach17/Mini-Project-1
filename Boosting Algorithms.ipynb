```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
X = np.loadtxt('train_2008.csv', skiprows = 1, delimiter = ',')
data = X[:, 3:-1]
labels = X[:, -1]
```

```python
# Adaboost
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size = 0.2)
clf = AdaBoostClassifier(n_estimators = 100)
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
# Gradient Boost
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(n_estimators = 100)
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
# XGBoost
from xgboost import XGBClassifier
clf = XGBClassifier()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
# XGBoost
from xgboost import XGBClassifier
clf = XGBClassifier(n_estimators = 200, silent = False)
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
# XGBoost
from xgboost import XGBClassifier
clf = XGBClassifier(n_estimators = 300, silent = False)
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
clf = XGBClassifier(n_estimators = 400)
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
clf = XGBClassifier(n_estimators = 500, max_depth = 4)
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
# Make prediction
data = X[:, 3:-1]
labels = X[:, -1]

X_out = np.loadtxt('test_2008.csv', skiprows=1, delimiter =',')
ids = X_out[:, 0]
X_out = X_out[:, 3:]


clf = XGBClassifier(n_estimators = 500)
clf.fit(data, labels)
predictions = clf.predict_proba(X_out)
prediction = np.vstack((ids, predictions[:, 1])).T
np.savetxt('predictionXGB500.csv', prediction, fmt='%4f', delimiter = ',')
```

```python
clf2 = XGBClassifier(n_estimators = 600, max_depth = 5)
clf2.fit(data, labels)
predictions = clf2.predict_proba(X_out)
prediction = np.vstack((ids, predictions[:, 1])).T
np.savetxt('predictionXGB600depth5.csv', prediction, fmt='%4f', delimiter = ',')
```
